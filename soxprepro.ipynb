```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/jaehyungyoo/Desktop/rltrader-master/sox.csv')
data.head()
```

```python
data['close'] = data['close'].str.replace(',','')
data['close_ma5'] = data['close_ma5'].str.replace(',','')
data['close_ma20'] = data['close_ma20'].str.replace(',','')
data['close_ma60'] = data['close_ma60'].str.replace(',','')
data['close_ma120'] = data['close_ma120'].str.replace(',','')
data.head()
```

```python
data['close'] = pd.to_numeric(data.close)
data['close_ma5'] = pd.to_numeric(data.close_ma5)
data['close_ma20'] = pd.to_numeric(data.close_ma20)
data['close_ma60'] = pd.to_numeric(data.close_ma60)
data['close_ma120'] = pd.to_numeric(data.close_ma120)
data.head()
```

```python
def preprocess(data):
    windows = [5, 20, 60, 120]
    for window in windows:
        data['close_ma%d_ratio' % window] = \
            (data['close'] - data['close_ma%d' % window]) \
            / data['close_ma%d' % window]
        
    return data
```

```python
data = preprocess(data)
data.head()
```
