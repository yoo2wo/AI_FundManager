```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/jaehyungyoo/Desktop/rltrader-master/exchange_rate.csv')
data.head()
```

```python
data['date'] = data['date'].str.replace('-','')
data.head()
```

```python
data['close'] = data['close'].str.replace(',','')
data.head()
```

```python
data['close'] = pd.to_numeric(data.close)
data.head()
```

```python
def preprocess(data):
    windows = [5, 20, 60, 120]
    for window in windows:
        data['close_ma{}'.format(window)] = \
            data['close'].rolling(window).mean()
        data['close_ma%d_ratio' % window] = \
            (data['close'] - data['close_ma%d' % window]) \
            / data['close_ma%d' % window]
    return data
```

```python
data = preprocess(data)
data.head()
```

```python
data.to_csv('/Users/jaehyungyoo/Desktop/rltrader-master/test.csv')
```
